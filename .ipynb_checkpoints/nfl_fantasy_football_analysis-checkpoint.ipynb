```python
%matplotlib inline

import pymysql
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import scale
from IPython.display import display, HTML
```

```python
conn = pymysql.connect(host='104.196.5.55', user='root', passwd='', db='nfl')
cur = conn.cursor()
```

```python
def query_sql(tbl_name):
    strSQL = ("Select * from " + tbl_name)
    df = pd.DataFrame(pd.read_sql(strSQL,conn,params=None))
    return df

#Clean this up with dictionary
df_td = query_sql("td")
df_pass = query_sql("pass")
df_offense = query_sql("offense")
df_play = query_sql("play")
df_player = query_sql("player")

df_offense = df_offense[['uid','gid','player','py','ints','tdp','ra','sra','ry','tdr','trg','rec','recy','tdrec','ret','rety','tdret','fuml','conv']]
df_player = df_player[['player','pname','pos1']]
df_play = df_play[['gid','pid']]
```

```python
#Inner join pass table and td table on play id to get passer
#Keep pid, yds_y, psr, trg
df_pass_td = pd.merge(df_td,df_pass,how='inner',on='pid')

#df_pass_td should have all plays that are tds, we subset to only get td passes 40 yards and over

df_pass_td = df_pass_td[['pid','yds_y','psr','trg']]
df_pass_td = df_pass_td[df_pass_td['yds_y']>39]

#Merge new pass td table with play table to get gid
df_pass_td_gid = pd.merge(df_pass_td,df_play,how='left',on='pid')

#Create binary for 40 yard and 50 yard td
df_pass_td_gid['40_yd_td'] = df_pass_td_gid['yds_y'].map(lambda x: 1 if 40 <= x else 0)
```

```python
df_offense_player = pd.merge(df_offense,df_player,how='inner',on='player')
```

```python
qb40 = pd.DataFrame(df_pass_td_gid.groupby(['gid','psr'])['40_yd_td'].agg(np.sum))
qb40 = qb40.add_suffix('_Count').reset_index()
df_offense = pd.merge(df_offense_player,qb40,how='left',left_on=['player','gid'],right_on=['psr','gid'])
df_offense['40_yd_td_Count'] = df_offense['40_yd_td_Count'].fillna(0)
```
